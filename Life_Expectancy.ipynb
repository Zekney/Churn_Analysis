```python
import matplotlib.pyplot as plt
```

```python
plt.Figure(figsize=(15,6))
plt.scatter(LE['Birth rate'], LE['Death rate'], color='Green')
plt.title('Birth Rate vs Death Rate')
plt.xlabel('Birth Rate (%)')
plt.ylabel('Death Rate (%)')
plt.grid(True)
plt.show
```

```python
plt.figure(figsize=(12,6))
plt.bar(LE['Country'], LE["Life expectancy males"],color='blue', label='Females',alpha=0.6)
plt.xticks(rotation=90)
plt.title('Life Expectancy by Country  (Males vs Females)')
plt.xlabel('Country')
plt.ylabel('Life Expectancy (Years)')
plt.legend()
plt.tight_layout()
plt.show
```
