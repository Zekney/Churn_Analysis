```python
#9 Data cleaning 
# Drop identifier column (they don't add analytical value)
DUP_clean = DUP.drop(columns=['RowNumber', 'CustomerId', 'Surname'])

# Check for duplicates
print('Duplicates:', DUP_clean.duplicated().sum())

# Ensure no missing values remain
DUP_clean.info()
```


Data Exploration

```python
#10 Univariate Analysis (Histograms)
numeric_cols = ['CreditScore','Age','Tenure','Balance','NumOfProducts','EstimatedSalary']

for col in numeric_cols:
    plt.figure(figsize=(6,3))
    plt.hist(DUP_clean[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('count')
    plt.tight_layout()
    plt.show()
```
